### Step 2: Define Colang Rails
We define a standard 'topical rail' to prevent the bot from discussing politics.


```python
colang_config = """
define user ask politics
  "What are your political views?"
  "Who should I vote for?"
  "Let's talk about elections."

define bot answer politics
  "I am an AI assistant focused on technical support. I cannot discuss politics."

define flow politics
  user ask politics
  bot answer politics

define flow greeting
  user express greeting
  bot express greeting
"""
```

### Step 3: Run the Guardrails Engine


```python
from nemoguardrails import LLMRails, RailsConfig

# We will use a mock LLM for local evaluation without API keys
yaml_config = """
models:
  - type: main
    engine: dummy
"""

config = RailsConfig.from_content(colang_content=colang_config, yaml_content=yaml_config)
rails = LLMRails(config)

# Test the topical rail
async def test_rail():
    response = await rails.generate_async(messages=[{'role': 'user', 'content': 'Who should I vote for in the next election?'}])
    print('Response:', response['content'])

import asyncio
asyncio.run(test_rail())
```
