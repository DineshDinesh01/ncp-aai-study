### Step 2: Initialize NIM Client
Set your `NVIDIA_API_KEY` to access models like `meta/llama-3.1-70b-instruct`.


```python
import os
from langchain_nvidia_ai_endpoints import ChatNVIDIA

# Ensure your API key is set in the environment
os.environ['NVIDIA_API_KEY'] = 'nvapi-YOUR_KEY_HERE'

llm = ChatNVIDIA(model='meta/llama-3.1-70b-instruct', temperature=0.0)
print('NIM initialized successfully!')
```

### Step 3: Define Tools & Build the ReAct Loop
We will give the agent a search tool to answer real-world questions.


```python
from langchain_core.tools import tool
from langchain.agents import create_tool_calling_agent, AgentExecutor
from langchain_core.prompts import ChatPromptTemplate

@tool
def get_nvidia_stock_price(ticker: str) -> str:
    """Gets the current stock price for NVIDIA."""
    # Mock function for lab purposes
    return 'NVIDIA (NVDA) stock price is roughly $135.20 USD as of the latest market close.'

tools = [get_nvidia_stock_price]
defined_prompt = ChatPromptTemplate.from_messages([
    ('system', 'You are a helpful financial assistant.'),
    ('human', '{input}'),
    ('placeholder', '{agent_scratchpad}')
])

agent = create_tool_calling_agent(llm, tools, defined_prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# Run the agent
response = agent_executor.invoke({'input': 'What is the current NVIDIA stock price?'})
print('\nFinal Answer:', response['output'])
```
